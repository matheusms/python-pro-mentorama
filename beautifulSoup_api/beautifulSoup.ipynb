```python
from requests import get
#raspagem do site 
resposta = get("https://pt.wikipedia.org/wiki/Clube_de_Regatas_do_Flamengo")
```

```python
from bs4 import BeautifulSoup
```

```python
tags = BeautifulSoup(resposta.text, 'html5lib')
```

```python
#pegando o nome da pagina
title = tags.find("title")
print("Página principal: ", title.text)
```

```python
#pegando todos os links
sub_sites = tags.find_all('a')
sub_sites
```

```python
links = ["https://pt.wikipedia.org" + k.get('href') for k in sub_sites if k.get('href') != None and k.get('href')[0] == '/']
links
```

```python
#realizando a raspagem dos sites
sites = []
for pv in links[0:10]:#diminui o range apenas para diminuir o tempo de processamento e facilitar o teste
    sites.append(get(pv))#ficará pesado para realizar
    
```

```python
#imprimir nome de cada pagina
for nomes in sites:
    tags2 = BeautifulSoup(nomes.text, 'html5lib')
    title = tags2.find("title")
    print("Página Secundária: ", title.text)
```
